#WEB SCRAPPING

Import some of basic libraries of web scrapping like beautifulsoup and need to clean the data using NLTK library

```python
import numpy as np
import pandas as pd
import requests as r
import bs4 as bf
from urllib.request import urlopen,urlparse, Request,HTTPError
from bs4 import BeautifulSoup as bfs
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
url="https://dir.indiamart.com/impcat/engineeredboards.html"
```

```python
link=r.get(url)
link
```

```python
html=bfs(link.content,"html.parser")
html
```

```python
heading = html.find_all(attrs={"class":"r-cl b-gry"})
type(heading)
```

```python
# company_name=[]
company_name=(html.findAll(attrs={"data-click":"^CompanyName"}))
company_name
```

```python
df = pd.DataFrame(company_name)
df
```

```python
df=df.rename(columns={0:"Seller_name"})
df
```

```python
phone_number=(html.find_all(attrs={"class":"pns_h duet"}))
phone_number
```

```python
df1 = pd.DataFrame(phone_number)
df1
```

```python
df1=df1.rename(columns={0:"Phone_number"})
df1
```

```python
Address=(html.find_all(attrs={"class":"to-txt"}))
Address
```

```python
df2= pd.DataFrame(Address)
df2
```

```python
df2=df2.rename(columns={0:"Address"})
df2
```

```python
df2["Address"]=df2["Address"].str.replace("\n","")
```

```python
data= pd.concat([df, df1,df2], axis=1)
data
```

```python
from google.colab import files
data.to_csv('data.csv')
files.download("data.csv")
```
